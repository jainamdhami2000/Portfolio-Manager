```python
import difflib
df1['symbol'] = ''
for x in df1['COMPANY_NAME']:
    closematch = difflib.get_close_matches(x.lower(), df2['Name'].str.lower(),1,cutoff=0.919)
#     df1[df1.loc[df1['COMPANY_NAME'] == x],2] = df2[df2.loc[df2['Name']==difflib.get_close_matches(x, df2['Name'])],0]
    if len(closematch)!=0:
        for i in df1.index[df1['COMPANY_NAME']==x].tolist():
            for j in df2.index[df2['Name'].str.lower()=="".join(closematch).lower()].tolist():
                df1.at[i,'symbol'] = df2.at[j,'Symbol']
```
